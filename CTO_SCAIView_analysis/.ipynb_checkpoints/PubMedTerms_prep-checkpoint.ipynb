```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page)
content = soup.body.find('div', attrs={'class':'article-page'}).text
```

```python
#content type is str
print(content)
```

```python
def list_to_string(list_name):
    newlist =';'.join(list_name)
    #print(newlist)
    return newlist

def find_PubMed_element(element):
    if line.startswith(element):
        cleanline = line.replace(element,'').replace('-','').lstrip()
        return cleanline
    

PM_terms_df = pd.DataFrame(columns = ['PMID', 'PT', 'OT','MH'])
content_list = content.splitlines()
PT_list = []
OT_list = []
MH_list = []
for line in content_list:
    if line.startswith('PMID'):
        PMID = line.replace(r'PMID','').replace('-','').lstrip()
    if line.startswith('PT'):
        PT = line.replace(r'PT','').replace('-','').lstrip()
        PT_list.append(PT)
    if line.startswith('OT'):
        OT = line.replace(r'OT','').replace('-','').lstrip()
        OT_list.append(OT)
    if line.startswith('MH'):
        MH = line.replace(r'MH','').replace('-','').lstrip()
        MH_list.append(MH)

PT_list_str = list_to_string(PT_list)
OT_list_str = list_to_string(OT_list)
MH_list_str = list_to_string(MH_list)

new_row = {'PMID':PMID, 'PT':PT_list_str, 'OT':OT_list_str, 'MH':MH_list_str}
print(new_row)
PM_terms_df = PM_terms_df.append(new_row, ignore_index=True)
display(PM_terms_df)
```
